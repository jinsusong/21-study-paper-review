## Abstract

명명된 엔티티 인식(NER)은 사람, 위치, 조직 등과 같은 사전 정의된 의미 유형에 속하는 텍스트에서 엄격한 지정자에 대한 언급을 식별하는 작업이다. 

NER은 항상 질문 답변, 텍스트 요약 및 기계 번역과 같은 많은 자연어 응용 프로그램의 기초 역할을 합니다. 

초기 NER 시스템은 도메인별 기능과 규칙을 설계하는 데 있어 인간 공학 비용으로 우수한 성능을 달성하는 데 큰 성공을 거두었다. 

최근 몇 년 동안 비선형 처리를 통한 지속적인 실제 값 벡터 표현과 의미 구성에 의해 강화된 딥 러닝이 NER 시스템에 채택되어 최첨단 성능을 발휘하고 있다. 

본 논문에서, 우리는 NER을 위한 기존 딥 러닝 기술에 대한 포괄적인 리뷰를 제공한다. 

먼저 태그가 지정된 NER 말뭉치와 기성 NER 도구를 포함한 NER 리소스를 소개한다. 

그런 다음 입력, 컨텍스트 인코더 및 태그 디코더에 대한 분산 표현이라는 세 가지 축을 따라 분류법을 기반으로 기존 작업을 체계적으로 분류한다. 

다음으로, 우리는 새로운 NER 문제 설정 및 응용 분야에서 최근에 적용된 딥 러닝 기술에 대한 가장 대표적인 방법을 조사한다. 

마지막으로 독자들에게 NER 시스템이 직면한 과제를 제시하고 이 영역의 향후 방향을 간략히 설명한다.

### NER approaches 


- Traditional
    1. Rule-based approaches : 
        - domain-specific 사전, 구문-어휘 패턴을 기반으로 규칙을 설계
        - 도메인 고유 규칙과 불완전한 사전으로 높은 정밀도, 낮은 재현율 관찰
        - 다른 도메인으로 시스템을 이전 X 
    2. Unsupervised learning : 
        - 가장 대표적인 접근법은 클러스터링 
        - 컨텍스트 유사성을 기반으로 클러스터링된 그룹에서 NE를 추출
        - 대규모 코퍼스에서 계산된 통계를 사용해 NE 언급 추론
    3. Feature-based supervised learning
        - 지도학습을 적용하면 NER은 다중 클래스 분류나 시퀀스 레이블 지정 테스크가 됨.
        - 어노테이션된 데이터 샘플이 주어지면 feature들이 각 training 예시를 잘 나타내도록 설계됨 
        - 이후 기계학습 알고리즘을 사용해 데이터에서 유사함 패턴을 인식하는 모델을 학습함.

- New 
    1. Deep-learning based approaches 
        - Dee-learning 을 적용하는 이유
            1. 활성화함수 덕분에 입력에서 출력으로 비선형 변환이 쉽다.
            2. NER 기능의 설계를 간단하게 해줌.
            3. end-to-end 학습이 가능함.

1. Distributed representations for input 
    1. Word-level Representation
    - pre-trained word embeddings : google word2vec, stanford GloVe, Facebook fastText, SENNA..
    - training시에 pre-trained word embedding을 고정하거나 fine-tuning을 진행

    2. Character-level Representation
    - 접두사 및 접미사와 같은 명시적 하위 단어 수준 정보를 활용하는데 유용
    - out-of-vocabulary를 자연스럽게 처리함 즉 보이지 않는 단어(unseen)을 표현하고 형태소 수준의 규칙성에 대한 정보를 고유
    - 두가지 아키텍처 : CNN, RNN(기본적으로 LSTM or GRU)

    3. Hybrid Representation
    - 추가 정보 (지명 목록, 어휘 유사성, 언어적 종속성 및 시각적 특징)를 최종 representations of words에 포함 
    - 시스템의 일반성을 손상시키는 대가로 NER 성능이 향상

Summary of DL-based NER 
    - NER 시스템의 성공은 input representation에 크게 의존함
    - 사전학습모델에 fine-tuning을 진행하는 것이 NER의 새로운 패러다임으로 성능이 크게 향상됨 
    - 하지만 아직 새로운 문장 또는 단어가 DL기반 NER 모델에 어떻게 추가할지에 대한 방법이 없음.
        - NER 성능이 외부 지식으로 향상될 수 있음을 보여주지만, 비용이 많이들고, 학습에 부정적인 영향 

    - RNN의 경우, 현재단계의 입력이 이전 단계의 출력을 필요로 하기 때문에 속도의 개선이 필요
    - 최종적으로 사용자가 선택할 architecture는 데이터 및 도메인 작업에 따라 다름
        - 데이터가 많은 경우 : 처음부터 RNN을 사용하여 모델을 훈련하고, 상황에 맞는 언어 모델을 fine-tuning
        - 데이터가 적은 경우 : 미리 사전 훈련된 모델이 많기 때문에 특정 영역(의료, 소셜 미디어)의 경우 상황에 맞는 언어 모델을 fine-tuning

## CHALLENGES and FUTURE DIRECTIONS
